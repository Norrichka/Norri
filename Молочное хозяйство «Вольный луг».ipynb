```python
def align_arrays(y_true, *y_preds):
    # Находим минимальную длину среди всех массивов
    min_length = min(len(y_true), *[len(pred) for pred in y_preds])
    
    # Обрезаем массивы до минимальной длины
    y_true_aligned = y_true[:min_length]
    y_preds_aligned = [pred[:min_length] for pred in y_preds]
    
    return y_true_aligned, y_preds_aligned

def compute_metrics(y_true, *y_preds):
    mse_scores = []
    mae_scores = []
    rmse_scores = []
    
    y_true_aligned, y_preds_aligned = align_arrays(y_true, *y_preds) # Используем функцию для выравнивания массивов
    
    for y_pred in y_preds_aligned:
        mse = mean_squared_error(y_true_aligned, y_pred)
        mae = mean_absolute_error(y_true_aligned, y_pred)
        rmse = mse ** 0.5
        mse_scores.append(mse)
        mae_scores.append(mae)
        rmse_scores.append(rmse)
    
    return mse_scores, mae_scores, rmse_scores

def print_metrics(mse_scores, mae_scores, rmse_scores):
    for i, (mse, mae, rmse) in enumerate(zip(mse_scores, mae_scores, rmse_scores), 1):
        print(f"MSE модели {i}: {mse}")
        print(f"MAE модели {i}: {mae}")
        print(f"RMSE модели {i}: {rmse}")
        print("")

# Assuming y_test, y_pred, y_pred_second, y_pred_third are defined elsewhere
mse_scores, mae_scores, rmse_scores = compute_metrics(y_test, y_pred_first, y_pred_second, y_pred_third)
print_metrics(mse_scores, mae_scores, rmse_scores)
```

Напишем функцию которая будет перебирать некоторые из порогов,посчитает, а потом выведет лучший порог с лучшими показателями

```python
def compute_precision_recall(y_true, y_scores, thresholds):
    precision, recall, _ = precision_recall_curve(y_test, y_scores)
    
    precision_values = []
    recall_values = []
    
    for thresh in thresholds:
        idx = np.searchsorted(thresholds, thresh, side="left")
        precision_values.append(precision[idx])
        recall_values.append(recall[idx])
        
    return precision_values, recall_values

thresholds = [round(i,2) for i in np.linspace(0.5, 1, num=20, endpoint=False)]
y_scores = model_log.predict_proba(X_test_ohe)[:, 1]

precision_values, recall_values = compute_precision_recall(y_test, y_scores, thresholds)
print("thresholds:", thresholds)
print("Precision:", precision_values)
print("Recall:", recall_values)
```
